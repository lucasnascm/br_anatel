```
import pandas as pd
import os
from zipfile import ZipFile
```

```
#definição do caminho do arquivo zip
  #os dados estão disponíveis em https://www.anatel.gov.br/paineis/acessos/telefonia-fixa
pasta = '/content/drive/MyDrive/br_anatel/telefonia_fixa/input'
mobile = os.path.join(pasta, 'acessos_telefonia_fixa.zip')
```

```
#decodificando os arquivos
with ZipFile(mobile) as z:
  with z.open(f'Acessos_Telefonia_Fixa_Autorizadas.csv') as f:
    tf_aut = pd.read_csv(f,sep=';', encoding='utf-8')
```

```
with ZipFile(mobile) as z:
  with z.open(f'Acessos_Telefonia_Fixa_Concessionarias.csv') as f:
    tf_conc = pd.read_csv(f,sep=';', encoding='utf-8')
```

```
tf_aut
```

```
tf_conc
```

```
telefonia_fixa_microdados = tf_conc.append(tf_aut, ignore_index=True)
```

```
telefonia_fixa_microdados.rename(columns={'Ano': 'ano','Mês':'mes', 'Grupo Econômico':'grupo_economico', 'Empresa':'empresa',
                                      'CNPJ':'cnpj', 'Porte da Prestadora':'porte_empresa', 'UF':'sigla_uf', 'Município':'municipio',
                                      'Código IBGE Município':'id_municipio', 'Tipo de Outorga':'outorga', 'Tipo do Acesso':'tipo_acesso',
                                      'Acessos':'acessos'}, inplace=True)
```
